```python
import numpy as np

col_name1=['col1']
list1=[1, 2, 3]
array1 = np.array(list1)
print('array1 shape: ', array1.shape)

# DataFrame generation with numpy ndarray
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('DataFrame with id_array:\n', df_list1)

# DataFrame generation with numpy ndarray
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('DataFrame with id_ndarray:\n', df_array1)

```

```python
# p51 needs 3 columns
col_name2 = ['col1', 'col2', 'col3']

# 2x3 list and ndarray generation, transform it to DataFrame
list2 = [[1, 2, 3,],
         [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape: ', array2.shape)
df_list2 = pd.DataFrame(list2, columns=col_name2)
print('DataFrame with 2d list:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns=col_name2)
print('DataFrame with 2d_ndarray:\n', df_array2)
```

```python
# Dictionary to DataFrame
# Key: columns, Value: list or ndarray column data
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict=pd.DataFrame(dict)
print('DataFrame with Dictionary: \n', df_dict)
```

```python
# DataFrame -> List
list3=df_dict.values.tolist()
print('df_dict.vlues.tolist() type: ', type(list3))
print(list3)

# DataFrame -> Dic
dict3 = df_dict.to_dict()
print('\n df_dict.to_dict() type: ', type(dict3))
print(dict3)
```

## DataFrame column data set generation and update p53

```python
titanic_df['Age_0']=0  # new column Age_0 and assign '0'
titanic_df.head(3)
```

```python
# new Columns Series
titanic_df['Age_by_10']=titanic_df['Age']*10
titanic_df['Family_No']=titanic_df['SibSp'] + titanic_df['Parch']+1
titanic_df.head(3)
```

```python
titanic_df['Age_by_10']=titanic_df['Age_by_10']+100
titanic_df.head(3)
```

```python
# Remove Columns
titanic_drop_df = titanic_df.drop('Age_0', axis=1) # axis0= row remove, axis1=column remove 
titanic_drop_df.head(3)
```

```python
# original data
titanic_df.head(3)
```

```python
drop_result=titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
print(' inplace=True drop return value: ', drop_result)
titanic_df.head(3)
```
